```python
from tkinter import*
from tkinter.ttk import Radiobutton
window=Tk()
window.title('Quiz Maker')
window.geometry('500x700')







label=Label()
label1=Label(window,text='1.Which organ covers the entire body and protects it?')
label1.pack()
selected1=StringVar()
a=Radiobutton(window,text='Liver',value='Liver',variable=selected1)
b=Radiobutton(window,text='Heart',value='Heart',variable=selected1)
c=Radiobutton(window,text='Skin',value='Skin',variable=selected1)
d=Radiobutton(window,text='Brain',value='Brain',variable=selected1)
a.place(x=130,y=50)
b.place(x=270,y=50)
c.place(x=130,y=100)
d.place(x=270,y=100)

selected2=StringVar()
label1=Label(window,text='2.What part of the body helps you move?')
label1.place(x=106,y=150)
a=Radiobutton(window,text='Eyes',value='Eyes',variable=selected2)
b=Radiobutton(window,text='Lungs',value='Lungs',variable=selected2)
c=Radiobutton(window,text='Pancrease',value='Pancrease',variable=selected2)
d=Radiobutton(window,text='Muscles',value='Muscles',variable=selected2)
a.place(x=130,y=200)
b.place(x=270,y=200)
c.place(x=130,y=250)
d.place(x=270,y=250)

selected3=StringVar()   
label1=Label(window,text='3.Which cell organelle is also called the powerhouse of the cell?')
label1.place(x=106,y=310)
a=Radiobutton(window,text='Ribosome',value='Ribosome',variable=selected3)
b=Radiobutton(window,text='Endoplasmic reticulam',value='Endoplasmic reticulam',variable=selected3)
c=Radiobutton(window,text='Cytoplasm',value='Cytplasm',variable=selected3)
d=Radiobutton(window,text='Mitochondria',value='Mitochondria',variable=selected3)
a.place(x=130,y=350)
b.place(x=270,y=350)
c.place(x=130,y=390)
d.place(x=270,y=390)


selected4=StringVar()
label1=Label(window,text='4.Where does our food get collected after we chew and swallow it?')
label1.place(x=106,y=450)
a=Radiobutton(window,text='Small intenstine',value='Small intenstine',variable=selected4)
b=Radiobutton(window,text='Liver',value='Liver',variable=selected4)
c=Radiobutton(window,text='Stomach',value='Stomach',variable=selected4)
d=Radiobutton(window,text='Large intenstine',value='Large intenstine',variable=selected4)
a.place(x=130,y=490)
b.place(x=270,y=490)
c.place(x=130,y=530)
d.place(x=270,y=530)
  


label=Label(window,text='Total mark',font=('Arial bold',10))
label.place(x=115,y=600)



    

a=['Skin','Muscles','Mitochondria','Stomach']

def result():
    count=0
    try:
        if a[0]==selected1.get():
             count=count+1
        if a[1]==selected2.get():
            count=count+1
        if a[2]==selected3.get():
            count=count+1
        if a[3]==selected4.get():
            count=count+1
    finally:
        entry1.insert(0,count)
        button1.config(state='disabled')
    
button1=Button(window,text='submitt',command =result)
button1.place(x=220,y=550)

entry1=Entry(window)
entry1.place(x=190,y=600)


window.mainloop()
```
